```python
#code snippet 2
#Importing h2o
import h2o

#initialize h2o
h2o.init()
```

```python
# Code snippet 3
# Importing dataset
data = h2o.import_file('https://raw.githubusercontent.com/Imjuandiaz/churn-data/refs/heads/main/WA_Fn-UseC_-Telco-Customer-Churn.csv')


data
```

```python
#code snippet 4
#Importing Kmeans Estimator
from h2o.estimators.kmeans import H2OKMeansEstimator
```

```python
#code snippet 5
#setting the number of clusters in the data train as 4
kmeans_model = H2OKMeansEstimator(k=6)

#Train the clustering model
kmeans_model.train(training_frame=data)
```

```python
#code snippet 6
#Displaying the cluster size
kmeans_model.size()
```

```python
#code snippet 7
#importing pandas
import pandas as pd

# predicts clusters on the rows
pred = kmeans_model.predict(data)

# join to original df
data_with_clusters = data.cbind(pred)

# import to pandas
df = data_with_clusters.as_data_frame()

# setting numerical variable: getting mean by cluster
numeric_summary = df.groupby("predict")[["tenure", "MonthlyCharges", "TotalCharges"]].mean()

# setting categorical variable: getting mean by cluster
categorical_cols = ["gender", "SeniorCitizen", "Partner", "Dependents",
                    "PhoneService", "MultipleLines", "InternetService",
                    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
                    "TechSupport", "StreamingTV", "StreamingMovies",
                    "Contract", "PaperlessBilling", "PaymentMethod"]

categorical_summary = df.groupby("predict")[categorical_cols].agg(lambda x: x.mode()[0])
```

```python
# combine resume
cluster_profiles = numeric_summary.join(categorical_summary)

# importing csv for Clusters caracteristics
cluster_profiles.to_csv("cluster_profiles.csv")

print(cluster_profiles)
```

```python
#code snippet 1
#Installing pyod package for anormaly detection algorithms (knn)
!pip install pyod
```

```python
#code snippet 2
#Importing pyod
from pyod.models.knn import KNN
```

```python
# Code snippet 3
# Importing the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Imjuandiaz/churn-data/refs/heads/main/WA_Fn-UseC_-Telco-Customer-Churn.csv')


data
```

```python
# Code snippet 4
# Create new dataset with only numerical columns
numeric_dataset = data.select_dtypes(include=['float64', 'int64'])
numeric_dataset
```

```python
# Code snippet 5
# Training with the KNN algorithm to identify the anomalies
ad_model = KNN()
ad_model.fit(numeric_dataset)
```

```python
# Code snippet 6
#Predict anomalies (1 = normal, -1 = Anomaly)
anomaly_labels = ad_model.labels_
anomaly_scores = ad_model.decision_scores_
```

```python
# Code snippet 7
# Adding the anomaly results to the original dataframe
data['Anomaly_Label'] = anomaly_labels
data['Anomaly_Score'] = anomaly_scores
data
```

```python
# Code snippet 8
# Exporting the dataset to a csv file with anomaly info
data.to_csv('Churn_KNN_Anomalies.csv', index=False)
```

```python
# Code snippet 9
# Displaying the number of anomalies
num_anomalies = (data['Anomaly_Label'] == 1).sum()
print(f"Number of anomalies: {num_anomalies}")
```
